```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (2+np.sin(x+2))*(x+2)-5
```

```python
x=np.linspace(-5,5,100)
y=f(x)

plt.plot(x,y)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('f(x)')
plt.grid()
```

```python
def bisect(f, a, b, tol=1e-6, max_iter=1000):
    intervals = []
    for i in range(max_iter):
        intervals.append((a, b))
        c = (a + b) / 2
        if f(c) == 0 or (b - a) / 2 < tol:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return c, intervals
```

```python
x_0,intervalls = bisect(f, -2, 2)
x_0
```

```python
plt.figure()

for a,b in intervalls[:5]:
    plt.plot(x_0, f(x_0), 'ro')
    plt.plot(x,y)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.title('f(x)')
    plt.grid()
    plt.axvline(x=a, color='g')
    plt.axvline(x=b, color='g')
    plt.show()
```

```python
def regula_falsi(f, a, b, tol=1e-6, max_iter=1000):
    for i in range(max_iter):
        c = (a * f(b) - b * f(a)) / (f(b) - f(a))
        if f(c) == 0 or (b - a) / 2 < tol:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return c
```

```python
x_0 = regula_falsi(f, 3, 5)
x_0
```

```python
plt.plot(x_0, f(x_0), 'ro')
plt.plot(x,y)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('f(x)')
plt.grid()
```

```python
def secant(f, x0, x1, tol=1e-6, max_iter=1000):
    for i in range(max_iter):
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if f(x2) == 0 or abs(x2 - x1) < tol:
            break
        x0 = x1
        x1 = x2
    return x2
```

```python
x_0 = secant(f, 1, 3)
x_0
```

```python
plt.plot(x_0, f(x_0), 'ro')
plt.plot(x,y)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('f(x)')
plt.grid()
```

```python
def newton_raphson(f, df, x0, tol=1e-6, max_iter=1000):
    for i in range(max_iter):
        x1 = x0 - f(x0) / df(x0)
        if f(x1) == 0 or abs(x1 - x0) < tol:
            break
        x0 = x1
    return x1
```

```python
def df(x):
    return (2+np.sin(x+2))+2*np.cos(x+2)
```

```python
x_0 = newton_raphson(f, df, 1)
x_0
```

```python
plt.plot(x_0, f(x_0), 'ro')
plt.plot(x,y)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('f(x)')
plt.grid()
```
