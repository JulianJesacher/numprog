```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 6-3*x

def fprime(x):
    return -3
```

```python
def g(x):
    return 1/(x-2)

def gprime(x):
    return -1/((x-2)**2)
```

```python
def condition(x, f, fprime):
    return np.abs(x*fprime(x)/(f(x)))
```

```python
x = np.linspace(0,3,250)

fY = f(x)
condF = condition(x, f, fprime)

gY = g(x)
condG = condition(x, g, gprime)
```

```python

plt.plot(x,fY)
plt.plot(x,condF)

plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','cond(f(x))'])
plt.grid()

plt.title('Kondition von f(x)=6-3x')
plt.show()
```

```python

plt.plot(x,gY)
plt.plot(x,condG)

plt.xlabel('x')
plt.ylabel('y')
plt.legend(['g(x)','cond(g(x))'])
plt.grid()

plt.ylim(-20,20)

plt.title('Kondition von g(x)=1/(x-2)')
plt.show()
```

# Wichtige Stellen

Besonders Polstellen und Nullstellen der Funktion an Stellen mit $x \neq 0$ sind wichtig.
Dort kann es sein dass die Kondition sehr groß wird.
